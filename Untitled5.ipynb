```python
import folium
import pandas as pd
```

```python
my_map1 = folium.Map(location = [22.30943, 114.231268], zoom_start = 13, tiles="cartodb positron" )
                                        
# Save method of Map object will create a map --> my_map
```

```python
# Read the csv file of OHCA patients

ohca_loc = pd.read_csv('/Users/doris/Desktop/FYP/OHCA.csv')
locations = list(zip(ohca_loc.lat, ohca_loc.lng))
```

```python
for i, r in ohca_loc.iterrows():
    folium.map.Marker(
        location=[r['lat'], r['lng']], 
        icon=folium.Icon(color="darkblue", icon="glyphicon glyphicon-user")
    ).add_to(my_map1)
my_map1
```
