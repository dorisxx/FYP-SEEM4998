```python
import folium
import pandas as pd
```

```python
my_map1 = folium.Map(location = [22.30943, 114.231268], zoom_start = 13, tiles="cartodb positron" )
                                        
# Save method of Map object will create a map --> my_map1
```

```python
# Read the csv file of candidate sites

candidate_loc = pd.read_csv('/Users/doris/Desktop/FYP/candidate_site.csv')
candidate_locations = list(zip(candidate_loc.lat, candidate_loc.lng))
```

```python
for i in range(len(candidate_locations)):
    folium.Circle(location=candidate_locations[i],radius=4, color='red').add_to(my_map1)
my_map1
```
